```julia
using SymPy

# Definimos las variables simbólicas
x, y = symbols("x y")

# Campo de velocidades v = (x, y)
v1 = x  # v_x
v2 = y  # v_y

# Calculamos la divergencia ∇ · v = ∂v_x/∂x + ∂v_y/∂y
div_v = diff(v1, x) + diff(v2, y)
```

```julia
using Plots

# Definimos la malla de puntos en el plano 2D
x = -2:0.5:2  # Rango de x, con pasos de 0.5
y = -2:0.5:2  # Rango de y, con pasos de 0.5
X = [xi for xi in x, yi in y]  # Malla de coordenadas x
Y = [yi for xi in x, yi in y]  # Malla de coordenadas y

# Calculamos las componentes del campo vectorial v = (-y, x)
U = X  # Componente v_x = -y
V = Y   # Componente v_y = x

# Graficamos el campo vectorial
quiver(X[:], Y[:], quiver=(U[:], V[:]), 
       title="Campo Vectorial v = (-y, x)", 
       xlabel="x", ylabel="y", 
       aspect_ratio=:equal)
```
